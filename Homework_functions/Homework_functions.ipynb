```python
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}

def num_search(doc_num):
    for doc in documents:
        if doc_num == doc['number']:
            return doc['name']       
        
def shelf_search(doc_num):
    for n, shelf in directories.items():
        if doc_num in shelf:
            return n  
        
def shelf_ad(shelf_num):
    if shelf_num not in directories.keys():
        directories[shelf_num] = []
        return directories.keys()

def shelf_del(shelf_num):
    if shelf_num in directories.keys() and not directories[shelf_num]:
        del(directories[shelf_num])
        return 'deleted' , directories.keys()
    elif shelf_num in directories.keys() and directories[shelf_num]:
        return 'not empty' , directories.keys()
    elif shelf_num not in directories.keys():
        return 'doesnt exist' , directories.keys()
    
def new_doc(doc_num , doc_type , doc_owner , doc_shelf):
    if doc_shelf in directories.keys():
        directories[doc_shelf].append(doc_num)
        document = {}
        document['type'] = doc_type
        document['number'] = doc_num
        document['name'] = doc_owner
        documents.append(document)     
        return documents
    
def doc_del(doc_num):
    for doc in documents:
        if doc_num == doc['number']:
            documents.remove(doc)
            return documents
        
def doc_rem(doc_num , doc_shelf):
    if doc_shelf not in directories.keys():
            return 'shelf doesnt exist' , directories.keys()
    v = False
    for n , shelf in directories.items():
            if doc_num in shelf:
                v = True
    if v == False:
            return 'doc doesnt exist' , documents
    for n, shelf in directories.items():
        if doc_num in shelf:
            shelf.remove(doc_num)
            break        
    directories[doc_shelf].append(doc_num)
    return 'done' , documents      
    
   
               
while True:
    user_input = input('Введите команду: ')
    if user_input == 'p':
        user_input1 = input('Введите номер документа: ')
        v = num_search(user_input1)
        if v == None:
            print('Документ не найден в базе')
        else: print('Владелец документа: ' , v)
    elif user_input == 's':
        user_input1 = input('Введите номер документа: ')
        s = shelf_search(user_input1)
        if s == None:
            print('Документ не найден в базе')
        else: print('Документ хранится на полке: ' , s)
    elif  user_input == 'l':
        user_input1 = doc['number']
        for doc in documents:
            print(f"№: {doc['number']} , тип: {doc['type']} , владелец: {doc['name']} , полка хранения: " , shelf_search(doc['number']))
    elif user_input == 'ads':
        user_input1 = input('Введите номер полки: ')
        ads = shelf_ad(user_input1)
        if ads == None:
            print(f"Такая полка уже существует. Текущий перечень полок: {', '.join(list(directories.keys()))}")
        else: print(f"Полка добавлена. Текущий перечень полок: {', '.join(list(ads))}") 
    elif user_input == 'ds':
        user_input1 = input('Введите номер полки: ')
        status , keys = shelf_del(user_input1)
        if status == 'deleted':
            print(f"Полка удалена. Текущий перечень полок: {', '.join(list(directories.keys()))}")
        elif status == 'not empty':
            print(f"На полке есть документы, удалите их перед удалением полки. Текущий перечень полок: {', '.join(list(directories.keys()))}")
        elif status == 'doesnt exist':
            print(f"Такой полки не существует. Текущий перечень полок: {', '.join(list(directories.keys()))}")
    elif user_input == 'ad':
        user_input1 = input('Введите номер документа: ')
        user_input2 = input('Введите тип документа: ')
        user_input3 = input('Введите владельца документа: ')
        user_input4 = input('Введите полку для хранения документа: ')
        v = new_doc(user_input1 , user_input2 , user_input3 , user_input4)
        if v == None:
            print(f"Такой полки не существует. Добавьте полку командой ads.\n")
        else:
            print('Документ добавлен.')
            
        print('Текущий список документов:')
        for doc in documents:
            print(f"№: {doc['number']} , тип: {doc['type']} , владелец: {doc['name']} , полка хранения: " , shelf_search(doc['number']))
    elif user_input == 'd':  
        user_input1 = input('Введите номер документа: ')
        d = doc_del(user_input1)
        if d == None:
            print(f"Документ не найден в базе.\n")
        else: print('Документ удалён.')
        print('Текущий список документов:')        
        for doc in documents:
            print(f"№: {doc['number']} , тип: {doc['type']} , владелец: {doc['name']} , полка хранения: " , shelf_search(doc['number']))
    elif user_input == 'm':
        user_input1 = input('Введите номер документа: ')
        user_input2 = input('Введите номер полки: ')
        status , keys = doc_rem(user_input1 , user_input2)
        if status == 'done':
            print('Документ перемещен.\nТекущий список документов:\n')
            for doc in documents:
                print(f"№{doc['number']} , тип: {doc['type']} , владелец: {doc['name']} , полка хранения: , {shelf_search(doc['number'])}")
        elif status == 'shelf doesnt exist':
            print(f"Такой полки не существует. Текущий перечень полок: {', '.join(list(directories.keys()))}") 
        elif status == 'doc doesnt exist':
            print('Документ не найден в базе.\nТекущий список документов:\n')
            for doc in documents:
                print(f"№{doc['number']} , тип: {doc['type']} , владелец: {doc['name']} , полка хранения: , {shelf_search(doc['number'])}")
    if user_input == 'q':
        break
    
```
