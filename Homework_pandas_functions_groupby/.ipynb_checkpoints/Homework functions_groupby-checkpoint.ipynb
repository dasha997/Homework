```python
import pandas as pd

def rating_class(row):
    if row <= 2:
        return 'low rate'
    elif row <= 4:
        return 'middle rate'
    elif row >= 4.5:
        return 'high rate'
    else: return 'unknown'

ratings = pd.read_csv('ratings.csv')
movies = pd.read_csv('movies.csv')
joined = ratings.merge(movies, on = 'movieId', how = 'left')
joined['class'] = joined['rating'].apply(rating_class)
joined.head(10)
```

```python
geo_data = {

'Центр': ['москва', 'тула', 'ярославль'],

'Северо-Запад': ['петербург', 'псков', 'мурманск'],

'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}

def geo_class(query):
    for k,v in geo_data.items():
        for v1 in v:
            if v1 in query.split(' '):
                return k
    return 'undefined'        
                        
queries = pd.read_csv('keywords.csv')
queries['region'] = queries['keyword'].apply(geo_class)
queries[queries['region'] != 'undefined'].head(10)
```
