```python
import pandas as pd
import numpy as np
horse_data = pd.read_csv('horse_data.csv', usecols = [2,3,4,10,12,19,22,6] , names = ['hospital_number','rectal_temperature','pulse','pain','abdominal_distension','total_protein','temperature_of_extremities','outcome'], na_values='?')

x = ['rectal_temperature','pulse','total_protein','temperature_of_extremities']

for colname in x:
    print(f"Базовые метрики для {colname}")
    print('размах', horse_data[colname].max() - horse_data[colname].min())
    print('мода', horse_data[colname].mode()[0])
    print('медиана', horse_data[colname].median())
    print('дисперсия', horse_data[colname].var())
    print('метод describe\n', horse_data[colname].describe())
    print('\n')

horse_data.head(10)
```

Согласно максимальным и минимальным значениям некоторых столбцов можно сделать вывод о наличии выбросов, т.к. максимальные показатели слишком велики, как, например, в pulse.

```python
for colname in x:
    q1 = horse_data[colname].quantile(0.25)
    q3 = horse_data[colname].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - (1.5 * iqr) 
    upper_bound = q3 + (1.5 * iqr)
    outliers = horse_data.loc[(horse_data[colname] < lower_bound)|(horse_data[colname] > upper_bound), colname]
    print(colname)
    print(outliers)
    print(f"Количество выбросов {outliers.count()}")   
```

 ## Анализ выбросов:
    Выбросы по ректальной температуре в пределах нормальных физиологических величин, поэтому ими можно пользоваться в расчётах, их не надо
    удалять или игнорировать. Они просто показывают, что лошадь больна.
    
    Выбросы по пульсу можно удалить, т.к. это странные показатели. Пульс лошади 24-42 уд/мин.
    
    В остальных двух столбцах выбросов нет,хотя в столбце  temperature_of_extremities они точно должны присутствовать, судя по минимальному и максимальному значениям.

Судя по всему, пропуски здесь полностью рандомные. Они не связаны с состоянием лошади. 

### rectal_temperature
пропуски в этом столбце можно заменить медианным значением, т.к. размах по этой величине небольшой, а пропусков очень много.
### pulse
предлагаю заполнить пропуски модальными значениями, тк оно ближе всего к норме. Медианное значение здесь уже повышенное.
### pain, total_protein и abdominal_distension
Поскольку данные по outcome наиболее полные, NaN в pain,total_protein и abdominal_distension будем заполнять медианами для соответствующих значений в outcome. В total_protein логично, что чем хуже белок, тем хуже состояние лошади.
### temperature_of_extremities
Удаляем столбец целиком, т.к. температура кала в 6 и 80 градусов маловероятна, таких данных там много и также присутствуют пропуски.
### outcome
всего один пропуск - можно взять моду.


```python
horse_data['outcome'].fillna(horse_data['outcome'].mode()[0], inplace = True)
horse_data['rectal_temperature'].fillna(horse_data['rectal_temperature'].median(), inplace = True)
horse_data['pulse'].fillna(horse_data['pulse'].mode()[0], inplace = True)
horse_data.pain.fillna(horse_data.groupby(['outcome']).pain.transform('median'), inplace=True)
horse_data.total_protein.fillna(horse_data.groupby(['outcome']).total_protein.transform('median'), inplace=True)
horse_data.abdominal_distension.fillna(horse_data.groupby(['outcome']).abdominal_distension.transform('median'), inplace=True)
horse_data = horse_data.drop(['temperature_of_extremities'], axis =1)

horse_data.head(30)
```
