```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('sex-offenders.csv')
df.head()

minor_murders_age = df.loc[(df['VICTIM MINOR'] == 'Y')& (df['GENDER'] == 'MALE'), ['AGE']]
minor_murders_age
minor_murders_age.plot(kind='hist', 
                   bins=10,
                   title='Male offenders age',
                   xticks=range(0, 100,10))
```

### Гипотеза №1

Большинство сексуальных преступлений в отношении несовершеннолетних в Чикаго совершают мужчины в возрасте от 30 до 40 лет. При этом удивительно, что существует небольшая группа людей, совершивших преступление в возрасте за 90 лет.

```python
import numpy as np

df.groupby('RACE')['RACE'].unique()
MALE = [
    df.loc[(df['RACE'] == 'BLACK') & (df['GENDER'] == 'MALE'), 'RACE'].count(),
    df.loc[(df['RACE'] == 'ASIAN/PACIFIC ISLANDER') & (df['GENDER'] == 'MALE'), 'RACE'].count(),
    df.loc[(df['RACE'] == 'BLACK HISPANIC') & (df['GENDER'] == 'MALE'), 'RACE'].count(),
    df.loc[(df['RACE'] == 'UNKNOWN') & (df['GENDER'] == 'MALE'), 'RACE'].count(),
    df.loc[(df['RACE'] == 'WHITE') & (df['GENDER'] == 'MALE'), 'RACE'].count(),
    df.loc[(df['RACE'] == 'WHITE HISPANIC') & (df['GENDER'] == 'MALE'), 'RACE'].count()
]

FEMALE = [
    df.loc[(df['RACE'] == 'BLACK') & (df['GENDER'] == 'FEMALE'), 'RACE'].count(),
    df.loc[(df['RACE'] == 'ASIAN/PACIFIC ISLANDER') & (df['GENDER'] == 'FEMALE'), 'RACE'].count(),
    df.loc[(df['RACE'] == 'BLACK HISPANIC') & (df['GENDER'] == 'FEMALE'), 'RACE'].count(),
    df.loc[(df['RACE'] == 'UNKNOWN') & (df['GENDER'] == 'FEMALE'), 'RACE'].count(),
    df.loc[(df['RACE'] == 'WHITE') & (df['GENDER'] == 'FEMALE'), 'RACE'].count(),
    df.loc[(df['RACE'] == 'WHITE HISPANIC') & (df['GENDER'] == 'FEMALE'), 'RACE'].count()
]

index = np.arange(6)
width = 0.35
plt.figure(figsize=(20,10))
plt.axis([-1,6,0,700])
plt.title('Crimes by race and gender', fontsize=20)
plt.bar(index, MALE, width, color='b')
plt.bar(index + width, FEMALE, width, color='r')
plt.yticks(np.arange(0,700, step=50))
plt.xticks(index, ['BLACK','ASIAN/PACIFIC ISLANDER','BLACK HISPANIC','UNKNOWN','WHITE','WHITE HISPANIC'])
plt.legend(['MALE','FEMALE'],fontsize=15)
plt.show()

```

### Гипотеза №2

Большинство сексуальных преступлений в Чикаго совершают темнокожие мужчины. Среди женщин большая часть сексуальных преступлений также приходится на людей с тёмным цветом кожи.

```python
df['BMI'] = round((df['WEIGHT']/2.205) / ((df['HEIGHT']/100) / 3.281 + df['HEIGHT']%10 / 39.37)**2)
df['COUNT'] = 1
fat_offenders = df.loc[df['BMI'] < 55,['BMI','COUNT']].groupby('BMI').count().reset_index().loc[:,['BMI','COUNT']]

data = {'Body Mass Index':fat_offenders['COUNT']}

x = fat_offenders['BMI'].to_numpy()
plt.figure(figsize=(20,10))
plt.plot(x,fat_offenders['COUNT'])
plt.legend(['Body Mass Index'], loc=1, fontsize=20)
plt.title('Body Mass Index of Chicago sex offenders', fontsize=20)
plt.xlabel('BMI', fontsize=20)
plt.ylabel('Quantity of people', fontsize=20)
plt.text(43, 60, "16<BMI<18 - Underweight\n18.5<BMI<25 - Normal\n25<BMI<30 - Overweight\n30<BMI<35 - Obese Class I\n35<BMI<40 - Obese Class II\nBMI>40 - Obese Class III",fontsize=20)
plt.show()
```

### Гипотеза №3

Большинство сексуальных маньяков в Чикаго имеют нормальный вес.

```python
Male_percent = len(df.loc[df['GENDER'] == 'MALE'])/1079
Female_percent = len(df.loc[df['GENDER'] == 'FEMALE'])/1079

plt.figure(figsize=(10,10))
labels = ['Male sex offenders','Female sex offenders']
values = [Male_percent,Female_percent]
colors = ['red','blue']
plt.pie(values,labels=labels,autopct='%1.1f%%',colors=colors)
plt.axis('equal')
plt.show()
```
